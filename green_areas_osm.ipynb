```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
# Specify the name that is used to seach for the data
place_name = " Toledo, Castile-La Mancha, Spain"
```

```python
# Fetch OSM street network from the location
graph = ox.graph_from_place(place_name)
```

```python
# Plot the streets
fig, ax = ox.plot_graph(graph)
```

```python
# Retrieve nodes and edges
nodes, edges = ox.graph_to_gdfs(graph)
```

```python
nodes.head()
```

```python
edges.head()
```

```python
# Get place boundary related to the place name as a geodataframe
area = ox.geocode_to_gdf(place_name)
area
```

```python
area.plot();
plt.axis('off');
```

```python
# List key-value pairs for tags
tags = {'building': True}
```

```python
buildings = ox.geometries_from_place(place_name, tags)
```

```python
buildings.head()
```

```python
# List key-value pairs for tags
tags = {'leisure': 'park', 'landuse': 'grass'}
```

```python
# Get the data
parks = ox.geometries_from_place(place_name, tags)

# Check the result
print("Retrieved", len(parks), "objects")
```

```python
parks.head()
```

```python
parks.plot(color="green");
plt.axis('off');
```

```python
fig, ax = plt.subplots(figsize=(12,8))

# Plot the footprint
area.plot(ax=ax, facecolor='black', zorder=0)

# Plot street edges
edges.plot(ax=ax, linewidth=0.5, edgecolor='dimgray', zorder=1)

# Plot buildings
buildings.plot(ax=ax, facecolor='silver', alpha=0.7, zorder=2)

# Plot restaurants
parks.plot(ax=ax, color='green', alpha=0.7, markersize=10, zorder=3)

plt.tight_layout()
plt.axis('off')
```
